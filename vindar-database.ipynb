```python
import duckdb
import pandas as pd

con = duckdb.connect() 
con.execute("INSTALL spatial; LOAD spatial;")

# Run your whole script
with open("database/schema_build.sql") as f:
    con.execute(f.read())
    
## Designation Analysis
## Distribution of missing-area percentage broken down by designation tier.
## Joins the missing area results to each parcel's top-tier PDO, then computes
## count and descriptive statistics (mean, median, sd, min, max) of
## area_missing_pct within each designation level + tier.

df = con.execute("""
    SELECT Designation_Level, tier, COUNT(*) AS n,
           ROUND(AVG(area_missing_pct),    2) AS avg_missing,
           ROUND(MEDIAN(area_missing_pct), 2) AS median_missing,
           ROUND(STDDEV(area_missing_pct), 2) AS sd_missing,
           ROUND(MIN(area_missing_pct),    2) AS min_missing,
           ROUND(MAX(area_missing_pct),    2) AS max_missing
    FROM area_missing a
    JOIN parcel_aoc p ON a.IDU = p.IDU
    GROUP BY Designation_Level, tier
    ORDER BY tier ASC
""").df()       

df
```
